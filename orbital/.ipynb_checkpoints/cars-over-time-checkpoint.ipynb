```python
import pandas as pd
import numpy as np
import sys
import os
sys.path.append("../xyme")
print(sys.path)
```

```python
folder = 'cars'
dfs = []
for fn in os.listdir(folder):
    df = pd.read_csv(os.path.join(folder, fn))
    df["date"] = df["date"].apply(pd.to_datetime)
    df["observed"] = (df["observed_or_modeled"] == "observed") * 1.0
    dfs.append(df)
all_df = pd.concat(dfs)
```

```python
sss = pd.read_csv("fcc_with_consensus_overtime.csv")
sss["report_date"] = sss["report_date"].apply(pd.to_datetime)
sss["consensus_date"] = sss["consensus_date"].apply(pd.to_datetime)
# sss[sss["target_value"].isna() & ((sss["report_date"].dt.year < 2019) | (sss["report_date"].dt.month <= 2))]
# sss["year"] = sss["report_date"].dt.year

# def quarter(df):
#     print(df)
#     return df

# sss.sort_values(["ticker", "report_date"]).groupby(["ticker", "year"]).apply(quarter)
sss.loc[sss["consensus_date"].isna(), "consensus_date"] = sss.loc[sss["consensus_date"].isna(), "report_date"]
sss = sss[~sss["reported"].isna()]
sss
```

```python
dt = sss[["ticker", "report_date", "consensus_date", "consensus", "reported"]]
dt = dt.rename(columns={"consensus_date": "date", "reported": "target", "report_date": "retrain_frequency"
                       }).set_index(["ticker", "date"])
full = all_df.set_index(["ticker", "date"])
full.sort_index(inplace=True)
dt.sort_index(inplace=True)
dt_t = dt[["retrain_frequency", "target"]]
dt_t = dt_t.reindex(labels=full.index, method="bfill")
dt_c = dt[["consensus"]]
dt_c = dt_c.reindex(labels=full.index, method="ffill")
dt_full = dt_t.join(dt_c).dropna()
dt_full.reset_index(inplace=True)
dt_full = dt_full[dt_full["date"] <= dt_full["retrain_frequency"]]
dt_full = dt_full.set_index(["ticker", "date"])
dt_full
```

## EPS Data

### Read EPS Data

```python
eps1 = pd.read_csv('EPS1.csv', parse_dates=['date'])
eps1.head()
```

```python
eps2 = pd.read_csv('EPS2.csv', parse_dates=['date'])
eps2.head()
```

```python
eps3 = pd.read_csv('EPS3.csv', parse_dates=['date'])
eps3.head()
```

```python
eps6 = pd.read_csv('EPS6.csv', parse_dates=['date'])
eps6.head()
```

```python
eps7 = pd.read_csv('EPS7.csv', parse_dates=['date'])
eps7.head()
```

### Unstacking all EPS Data

```python
trans_eps1 = eps1.set_index('date').unstack().reset_index()
trans_eps1.rename(columns={'level_0':'symbol', 0:'eps'}, inplace=True)
trans_eps1 = trans_eps1.sort_values(['symbol','date']).reset_index(drop=True)
```

```python
trans_eps1.head()
```

```python
trans_eps2 = eps2.set_index('date').unstack().reset_index()
trans_eps2.rename(columns={'level_0':'symbol', 0:'eps'}, inplace=True)
trans_eps2 = trans_eps2.sort_values(['symbol','date']).reset_index(drop=True)
```

```python
trans_eps2.head()
```

```python
trans_eps3 = eps3.set_index('date').unstack().reset_index()
trans_eps3.rename(columns={'level_0':'symbol', 0:'eps'}, inplace=True)
trans_eps3 = trans_eps3.sort_values(['symbol','date']).reset_index(drop=True)
```

```python
trans_eps3.head()
```

```python
trans_eps4 = eps4.set_index('date').unstack().reset_index()
trans_eps4.rename(columns={'level_0':'symbol', 0:'eps'}, inplace=True)
trans_eps4 = trans_eps4.sort_values(['symbol','date']).reset_index(drop=True)
```

```python
trans_eps4.head()
```

```python
trans_eps5 = eps5.set_index('date').unstack().reset_index()
trans_eps5.rename(columns={'level_0':'symbol', 0:'eps'}, inplace=True)
trans_eps5 = trans_eps5.sort_values(['symbol','date']).reset_index(drop=True)
```

```python
trans_eps5.head()
```

```python
trans_eps6 = eps6.set_index('date').unstack().reset_index()
trans_eps6.rename(columns={'level_0':'symbol', 0:'eps'}, inplace=True)
trans_eps6 = trans_eps6.sort_values(['symbol','date']).reset_index(drop=True)
```

```python
trans_eps6.head()
```

```python
trans_eps7 = eps7.set_index('date').unstack().reset_index()
trans_eps7.rename(columns={'level_0':'symbol', 0:'eps'}, inplace=True)
trans_eps7 = trans_eps7.sort_values(['symbol','date']).reset_index(drop=True)
```

```python
trans_eps7.head()
```

```python
# Append all the eps files together
all_eps = trans_eps1.append([trans_eps2, trans_eps3, trans_eps4, 
                            trans_eps5, trans_eps6, trans_eps7], ignore_index=True)
```

```python
# Check if the merge is correct
len(all_eps) == len(trans_eps1) + len(trans_eps2) + len(trans_eps3) + len(trans_eps4) + len(trans_eps5) + len(trans_eps6) + len(trans_eps7)
```

```python
# check if the stack is correct after merge
eps1.shape[1]+eps2.shape[1]+eps3.shape[1]+eps4.shape[1]+eps5.shape[1]+eps6.shape[1]+eps7.shape[1]-7
```

```python
all_eps.symbol.nunique()
```

```python
all_eps = all_eps.sort_values(['symbol', 'date']).reset_index(drop=True)
all_eps.head()
```

### Checking the Data Quality

```python
# Check the date gap
eps_test = all_eps.copy()
eps_test = eps_test.sort_values('date').reset_index(drop=True)
eps_test['next_date'] = eps_test.groupby('symbol')['date'].shift(-1)
eps_test['date_gap'] = eps_test.apply(lambda x:(x['next_date'] - x['date']).days, axis=1)
```

```python
eps_test[(eps_test['date_gap'].notnull()) & (eps_test['date_gap'] != 7)].sort_values('date_gap', ascending=False).head()
```

# Import

```python
import pandas as pd
from functools import reduce
```

## Read Data

```python
eps1=pd.read_csv('./E and P_S/EPS1.csv',parse_dates=['date']).sort_values('date').reset_index(drop=True)
eps2=pd.read_csv('./E and P_S/EPS2.csv',parse_dates=['date']).sort_values('date').reset_index(drop=True)
eps3=pd.read_csv('./E and P_S/EPS3.csv',parse_dates=['date']).sort_values('date').reset_index(drop=True)
eps4=pd.read_csv('./E and P_S/EPS4.csv',parse_dates=['date']).sort_values('date').reset_index(drop=True)
eps5=pd.read_csv('./E and P_S/EPS5.csv',parse_dates=['date']).sort_values('date').reset_index(drop=True)
eps6=pd.read_csv('./E and P_S/EPS6.csv',parse_dates=['date']).sort_values('date').reset_index(drop=True)
eps7=pd.read_csv('./E and P_S/EPS7.csv',parse_dates=['date']).sort_values('date').reset_index(drop=True)
```

## Merge All Files

```python
eps_list=[eps2,eps3,eps4,eps5,eps6,eps7]
eps=reduce(lambda l,r: l.merge(r,on=['date'],how='outer'),eps_list)
#eps=pd.concat(eps_list,join='outer', axis=1)
eps = eps.sort_values("date").reset_index(drop=True)
```

```python
len(eps.columns) - 1 == sum([len(df.columns) - 1 for df in eps_list])
```

```python
# Drop the data without any eps data
left_eps = eps.copy()
left_eps = left_eps.dropna(axis=1, how='all')
left_eps.head()
```

```python
trans_eps =left_eps.set_index("date").unstack().reset_index()
trans_eps.rename(columns={'level_0':'ticker',0:"next_twelve_months_eps"},inplace=True)
trans_eps=trans_eps.sort_values(['ticker','date']).reset_index(drop=True)
```

## Quality check

```python
eps_test=trans_eps.copy()
```

```python
eps_test['next_date']=eps_test.groupby('ticker').date.shift(-1)
```

```python
eps_test['gap']=eps_test.apply(lambda x: (x.next_date-x.date).days,axis=1)
```

# Import

```python
import pandas as pd
import numpy as np
from datetime import timedelta
```

# Read Data

```python
train = pd.read_csv('XYME_Train.csv')
test = pd.read_csv('XYME_Test.csv')
```

# Transformation 

```python
# the test data doesn't contain the target column (Y_Variable)
# add it in for later merge
test['Y_Variable'] = pd.Series()
# merge train and test data
after_merge = pd.concat([train, test], ignore_index=True, sort=False)
after_merge.head()
```

```python
# check the length of the new dataframe
len(train) + len(test) == len(after_merge)
```

```python
len(after_merge)
```

```python
train.shape[0]
```

```python
# since both train and test data don't have date column
# add dummy date to the second level
date_list = pd.date_range(end='2019-10-07', periods=after_merge.shape[0], freq='S').tolist()
after_merge.insert(after_merge.columns.get_loc('ID'), column='Dates', value=date_list)
```

```python
# add retrain frequency, retrain once per 1000 records
arr1 = []
np.asarray(arr1)
for i in range(train.shape[0] // 1000):
    temp = np.repeat(after_merge.Dates[(i+1) * 1000 - 1], 1000)
    arr1 = np.append(arr1, temp)
arr2 = np.repeat(after_merge.Dates[train.shape[0] - 1], (train.shape[0] - train.shape[0] // 1000 * 1000))
arr3 = np.repeat(after_merge.Dates[train.shape[0] - 1] + timedelta(seconds=1), test.shape[0])
freq_list = np.concatenate([arr1, arr2, arr3], axis=0)
after_merge.insert(after_merge.columns.get_loc('ID'), column='Retrain_Freq', value=freq_list)
```

# Output

```python
after_merge.to_csv('after_merge.csv', index=False)
```
